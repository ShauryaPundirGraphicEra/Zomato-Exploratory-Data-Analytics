```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
```

```python
orders = pd.read_csv('Orders.csv')
```

```python
orders
```

```python
orders.info()
```

```python
orders["Order_Date"]=pd.to_datetime(orders["Order_Date"])
```

```python
orders.info()
```

```python
orders
```

```python
orders["Month"]=orders["Order_Date"].dt.strftime("%B")
orders
```

```python
sns.countplot(x="Month",data=orders)
plt.xticks(rotation=90)
plt.show()
```

```python
orders["Hour"]=orders["Order_Date"].dt.strftime("%H")
orders
orders
```

```python
orders["Hour"]=orders["Hour"].astype(int)

# Sort the DataFrame by the 'Hour' column and assign the result to a temporary variable or use inplace=True
orders_sorted = orders.sort_values(by="Hour") # or orders.sort_values(by="Hour", inplace=True)


orders["Hour"] = orders_sorted["Hour"] # Skip this line if inplace=True is used

sns.lineplot(x="Hour",y="Total_Amount",data=orders,estimator="sum")
plt.xticks(rotation=90)
plt.show()
```

```python
customers=pd.read_csv("Customers.csv")
```

```python
df=pd.merge(left=orders,right=customers,on="Customer_ID",how="inner")
```

```python
df
```

```python
sns.countplot(x="Customer_Location",data=df,hue="Order_Status")
plt.xticks(rotation=90)
plt.show()
```

```python
plt.figure(figsize=(10,10))
sns.countplot(x="Customer_Location",data=df,hue="Customer_Age_Group",palette="viridis",edgecolor="black")
plt.legend(bbox_to_anchor=(1,1))
plt.show()
```
